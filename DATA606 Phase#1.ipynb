```python
print(data_new['Location'].unique())
```

```python
data_state = data_new[data_new["Location"].str.contains("PW|BP2|FM|VA2|US|IH2|DD2|MH|RP|LTC") == False]
data_state
```

```python
Distributed_Vaccines = data_state.groupby('Location')['Distributed'].sum()
Distributed_Vaccines
```

```python
Distributed_Vaccines_Moderna = data_state.groupby('Location')['Distributed_Moderna'].sum()
Distributed_Vaccines_Moderna
```

```python
Distributed_Vaccines_Pfizer = data_state.groupby('Location')['Distributed_Pfizer'].sum()
Distributed_Vaccines_Pfizer
```

```python
Distributed_Vaccines_Janssen = data_state.groupby('Location')['Distributed_Janssen'].sum()
Distributed_Vaccines_Janssen
```

```python
Percentage_Moderna = Distributed_Vaccines_Moderna / Distributed_Vaccines *100
print(Percentage_Moderna)
```

```python
Percentage_Pfizer = Distributed_Vaccines_Pfizer / Distributed_Vaccines *100
print(Percentage_Pfizer)
```
